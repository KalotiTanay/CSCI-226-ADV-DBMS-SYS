# Dependancies Install Here

```python
import os, getpass, pandas as pd
from sqlalchemy import create_engine, text
from google.colab import userdata
```

# Connect with DB

```python
print('Connecting with NeonDB...')
STUDENT_DB_URL = userdata.get('NEON_STUDENT_DB_URL') or getpass.getpass("Paste STUDENT DB URL (hidden): ")
engine = create_engine(STUDENT_DB_URL, pool_pre_ping=True, future=True)
print('Connected with NeonDB successfully!!!') if (engine) else print('Failed to establish connection!!!')
```

# Example 1: List out all laptop makers and their average laptop prices

```python
# Example 1: List out all laptop makers and their average laptop prices
with engine.connect() as conn:
    prices = """
    SELECT p.maker as "Maker", ROUND(AVG(l.price),2) AS "Price (Average)"
    FROM product p
    JOIN laptop l ON l.model = p.model
    GROUP BY Maker
    ORDER BY "Maker" ASC;
    """
    display('Average Laptop Prices by Makers', pd.read_sql_query(text(prices), conn))
```

# Example 2: Find all color printers and their makers

```python
# Example 2: Find all color printers and their makers
with engine.connect() as conn:
    printers = """
    SELECT p.maker as "Maker", pr.model as "Model", pr.type as "Type", pr.price as "Price"
    FROM product p
    JOIN printer pr ON pr.model = p.model
    WHERE pr.color = TRUE
    ORDER BY Price;
    """
    display('All Color Printers Offered by Price', pd.read_sql_query(text(printers), conn))
```

# Example 3: Create a Unified Catalog of all the products by Makers

```python
# Example 3: Create a Unified Catalog of all the products by Makers
with engine.connect() as conn:
    catalog = """
    SELECT
    p.maker, p.model, p.type,
    pc.speed AS pc_speed, pc.ram AS pc_ram, pc.hd AS pc_hd, pc.price AS pc_price,
    l.speed AS laptop_speed, l.ram AS laptop_ram, l.hd AS laptop_hd, l.screen AS laptop_screen, l.price AS laptop_price,
    pr.color AS printer_color, pr.type AS printer_type, pr.price AS printer_price

    FROM product p
    LEFT JOIN pc pc ON pc.model = p.model
    LEFT JOIN laptop l ON l.model = p.model
    LEFT JOIN printer pr ON pr.model = p.model
    ORDER BY p.maker, p.model;
    """
    display('Unified Catalog of All Products Offered by Makers', pd.read_sql_query(text(catalog), conn))
```
